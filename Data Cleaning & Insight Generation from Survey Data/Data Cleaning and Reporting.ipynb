# Create Dashboard for  highlights the insights 

```python
# Dashboard size
plt.figure(figsize=(18,12))

# Plot 1: Age Distribution
plt.subplot(2,3,1)
df2["Q1"].value_counts().plot(kind="bar", color="skyblue")
plt.title("Age Distribution")
plt.xlabel("Age Groups")
plt.ylabel("Count")

# Plot 2: Gender of Respondents
plt.subplot(2,3,2)
df2["Q2"].value_counts().plot(kind="bar", color="lightgreen")
plt.title("Gender of Respondents")
plt.xlabel("Gender")
plt.ylabel("Count")

# Plot 3: Top 10 Countries
plt.subplot(2,3,3)
df2["Q3"].value_counts().head(10).plot(kind="bar", color="orange")
plt.title("Top 10 Countries")
plt.xlabel("Countries")
plt.ylabel("Count")

# Plot 4: Programming Languages (Q7_Part_1)
plt.subplot(2,3,4)
lang_counts = df2["Q7_Part_1"].value_counts().head(10)
sns.barplot(x=lang_counts.values, y=lang_counts.index, orient='h', palette="Blues_r")
plt.title("Most Usable Programming Languages (Q7)")
plt.xlabel("Count")
plt.ylabel("Programming Languages")

# Plot 5: Programming Languages (Q6)
plt.subplot(2,3,5)
lang_counts = df2["Q6"].value_counts().head(10)
sns.barplot(x=lang_counts.values, y=lang_counts.index, orient='h', palette="Greens_r")
plt.title("Most Usable Programming Languages (Q6)")
plt.xlabel("Count")
plt.ylabel("Programming Languages")

# Adjust layout
plt.tight_layout()
plt.show()
```
